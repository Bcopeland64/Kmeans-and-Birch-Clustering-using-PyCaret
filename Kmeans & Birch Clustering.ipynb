```python
data_train = dataset.sample(frac=0.80, random_state=123)
```

```python
data_test = dataset.drop(data_train.index)
```

```python
data_train.head()
```

```python
data_test.reset_index(drop=True, inplace=True)
```

```python
data_test.head()
```

```python
from pycaret.clustering import *
```

```python
#ignore_feature holds out the label you want to predict
clust = setup(data=data_train, normalize=True, ignore_features=['MouseID'],
              session_id=123)
```

```python
kmeans = create_model('kmeans')
```

```python
print(kmeans)
```

```python
kmodes = create_model("kmodes", num_clusters=6)
```

```python
print(kmodes)
```

```python
models()
```

```python
kmeans_result = assign_model(kmeans)
```

```python
kmeans_result.head()
```

```python
#PCA Plot to more easily plot and visualize larger datasets
plot_model(kmeans)
```

```python
#Elbow plots allow us to find the optimal number of clusters (5 as is shown below)
plot_model(kmeans, plot='elbow')
```

```python
#Silhouette scores allow for the evaluation of of our clustering algorithm
plot_model(kmeans, plot='silhouette')
```

```python
#Distribution plot allows us to assess the the distribution of a sample comparing the empirical distribution of the data with the theoretical values expected from a specified distribution
plot_model(kmeans, plot='distribution')
```

```python
plot_model(kmeans, plot='distribution', feature='class')
```

```python
#Combination of TSNE plots, Elbow Plots, Silhouette, Distance, and Distribution for side-by-side comaparison
evaluate_model(kmeans)
```

```python
test_pred = predict_model(kmeans, data=data_test)
```

```python
test_pred.head()
```

```python
save_model(kmeans, "kmeans_model")
```

```python
saved_model = load_model('kmeans_model')
```
